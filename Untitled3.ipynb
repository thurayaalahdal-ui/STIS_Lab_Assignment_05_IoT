```python
from ecdsa import SigningKey, VerifyingKey, NIST256p, BadSignatureError
import hashlib
import binascii
sk = SigningKey.generate(curve=NIST256p)
vk = sk.verifying_key
print("Private key (PEM):\n", sk.to_pem().decode()[:200], "...")
print("Public key (PEM):\n", vk.to_pem().decode()[:200], "...")
message = "Ahmed pays 100 coins to Ali"
print("\nTransaction Message:", message)
h_hex = hashlib.sha256(message.encode()).hexdigest()
print("SHA-256 Hash:", h_hex)
signature = sk.sign(message.encode())
sig_hex = binascii.hexlify(signature).decode()
print("Signature (hex):", sig_hex)
print("Signature length (bytes):", len(signature))
try:
    verified = vk.verify(signature, message.encode())
    print("Signature Verified:", verified)
except BadSignatureError:
    print("Signature Verified: False (BadSignatureError)")
tampered_message = "Ahmed pays 1000 coins to Ali"
try:
    verified_tampered = vk.verify(signature, tampered_message.encode())
    print("\nSignature Verified on tampered message:", verified_tampered)
except BadSignatureError:
    print("\nSignature Verified on tampered message: False (tampering detected)")

```
